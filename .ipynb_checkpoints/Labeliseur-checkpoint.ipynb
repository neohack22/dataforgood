## Script à exécuter 

```python
import pandas as pd
import numpy as np
from os import path

comments = pd.read_csv('comments.csv')
words_1 = "fraiche|pourri|DLC|empoison|frai|moisi|périmé|hygién|sale|coll|mouche|propre|froid|chaine"
words_2 = "|fuir|netto|ventre|gast|poussi|pourr|vomi|malade|trace|intoxication|maux|fuir|évit"
words = words_1 + words_2
mask = comments['comments'].str.contains(words)
occurences = comments.loc[mask].index
comments = comments.loc[occurences]


def create_unlabeled_dataset(data):
    
    # Split of the work
    data = data.sample(n=1500, random_state=10)
    dataset_1 = data.iloc[0:500]
    dataset_2 = data.iloc[500:1000]
    dataset_3 = data.iloc[1000:1500]
    user = input("What is your name ? Type Abdoulai or Nisha or Adou.")
    if(user=="Abdoulai"):
        data = dataset_1
        data['label'] = np.nan
    elif(user=="Nisha"):
        data = dataset_2
        data['label'] = np.nan
    elif(user=="Adou"):
        data = dataset_3
        data['label'] = np.nan
    
    data.to_csv('unlabeled_data.csv', index=False)
    
    return data


def labelisation(data):
    
    if(path.isfile('unlabeled_data.csv')==False):
        unlabeled_data = create_unlabeled_dataset(data)
    else:
        unlabeled_data = pd.read_csv('unlabeled_data.csv')
        
    if(path.isfile('labeled_data.csv')==False):
        labeled_data = pd.DataFrame({'date':[], 'comments':[], 'restaurant':[], 'label':[]})
        labeled_data.to_csv('labeled_data.csv', index=False)
        
    count = len(unlabeled_data)
    
    while(count > 0):
        unlabeled_data = pd.read_csv('unlabeled_data.csv')
        comment = unlabeled_data['comments'].iloc[0]
        label = input(comment + "\n\nPercevez-vous un manque d'hygiène ?\n\nTapez '0' pour non ou tapez '1' pour oui.\n\n")
        
        # Update of the labeled dataframe
        labeled_data = pd.read_csv('labeled_data.csv')
        unlabeled_data['label'].iloc[0] = int(label)
        row = unlabeled_data.iloc[0]
        labeled_data = labeled_data.append(row)
        labeled_data.to_csv('labeled_data.csv', index=False)
        
        # Update of the unlabeled dataframe
        unlabeled_data = unlabeled_data.iloc[1:]
        unlabeled_data.to_csv('unlabeled_data.csv', index=False)

    return 'Done'
```

## Fonction à exécuter

```python
labelisation(comments)
```
